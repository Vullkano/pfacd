```python
import numpy as np
from pyprojroot.here import here
import pandas as pd
from pyarrow import feather
import janitor
from janitor import clean_names

"""
T_MED	ºC	Temperatura média do ar a 1,5 m 	
T_MAX	ºC	Temperatura máxima do ar a 1,5 m 	
T_MIN	ºC	Temperatura mínima do ar a 1,5 m 	
DD_MED	º	Rumo médio do vento	
DD_FFX	º	Rumo do vento máximo	
FF_MED	m/s	Intensidade média do vento a 10 m	
FF_MAX	m/s	Intensidade máxima instantânea do vento	
PR_QTD	mm	Quantidade de precipitação	

1200551	Viana do Castelo / Chafé	
1210622	Braga / Merelim	
1200567	Vila Real / Aeródromo	
1200575	Bragança	
1200545	Porto / Pedras Rubras	
1210702	Aveiro / Universidade	
1200560	Viseu / Centro Coordenador	
1210683	Guarda	
1200548	Coimbra / Cernache / Aeródromo	
1200570	Castelo Branco 	
1210718	Leiria	
1210734	Santarém / Fonte Boa  Est. Zootécnica	
1200571	Portalegre	
1200579	Lisboa / Gago Coutinho	
1210770	Setúbal / Estação de Fruticultura	
1200558	Évora / Aeródromo	
1200562	Beja	
1200554	Faro / Aeroporto	

"""

estacoes = {
    "Viana do Castelo": 1200551,
    "Braga": 1210622,
    "Vila Real": 1200567,
    "Bragança": 1200575,
    "Porto": 1200545,
    "Aveiro": 1210702,
    "Viseu": 1200560,
    "Guarda": 1210683,
    "Coimbra": 1200548,
    "Castelo Branco": 1200570,
    "Leiria": 1210718,
    "Santarém": 1210734,
    "Portalegre": 1200571,
    "Lisboa": 1200579,
    "Setúbal": 1210770,
    "Évora": 1200558,
    "Beja": 1200562,
    "Faro": 1200554
}
T_MED = pd.read_excel(here("data/dados_ipma/Dados_diarios_T_Vento_Prec_capitaisDistrito_2013 a 2023.xlsx"), sheet_name="T_MED")
T_MAX = pd.read_excel(here("data/dados_ipma/Dados_diarios_T_Vento_Prec_capitaisDistrito_2013 a 2023.xlsx"), sheet_name="T_MAX")
T_MIN = pd.read_excel(here("data/dados_ipma/Dados_diarios_T_Vento_Prec_capitaisDistrito_2013 a 2023.xlsx"), sheet_name="T_MIN")
DD_MED = pd.read_excel(here("data/dados_ipma/Dados_diarios_T_Vento_Prec_capitaisDistrito_2013 a 2023.xlsx"), sheet_name="DD_MED")
DD_FFX = pd.read_excel(here("data/dados_ipma/Dados_diarios_T_Vento_Prec_capitaisDistrito_2013 a 2023.xlsx"), sheet_name="DD_FFX")
FF_MED = pd.read_excel(here("data/dados_ipma/Dados_diarios_T_Vento_Prec_capitaisDistrito_2013 a 2023.xlsx"), sheet_name="FF_MED")
FF_MAX = pd.read_excel(here("data/dados_ipma/Dados_diarios_T_Vento_Prec_capitaisDistrito_2013 a 2023.xlsx"), sheet_name="FF_MAX")
PR_QTD = pd.read_excel(here("data/dados_ipma/Dados_diarios_T_Vento_Prec_capitaisDistrito_2013 a 2023.xlsx"), sheet_name="PR_QTD")
list_of_these = [T_MED, T_MAX, T_MIN, DD_MED, DD_FFX, FF_MED, FF_MAX, PR_QTD]
names = ["T_MED", "T_MAX", "T_MIN", "DD_MED", "DD_FFX", "FF_MED", "FF_MAX", "PR_QTD"]
T_MED
```

```python
for i in range(len(list_of_these)):
  for a, b in estacoes.items():
    list_of_these[i] = list_of_these[i].rename_column(b, a)
list_of_these[1]
```

```python
for i in range(len(list_of_these)):
  list_of_these[i] = list_of_these[i].melt(id_vars=["ANO", "MES", "DIA"], var_name="Estacao", value_name=names[i])
list_of_these[1]
```

```python
# join them
base = list_of_these[0]
for i in range(1, len(list_of_these)):
  base = base.merge(list_of_these[i], on=["ANO", "MES", "DIA", "Estacao"], how="outer")
base
```

```python
from skimpy import skim
# substitute where -999.0 for NA
base = base.replace(-990.0, np.nan)
skim(base)
```

```python
# date to timestamp
base["date"] = pd.to_datetime(base.rename_column("ANO", "year").rename_column("MES", "month").rename_column("DIA", "day")[["year", "month", "day"]])
base
```

```python
from janitor.timeseries import fill_missing_timestamps
groups = base.groupby("Estacao")
groups = [groups.get_group(x) for x in groups.groups]
for group in groups:
  group = fill_missing_timestamps(group, frequency="1D")
base_filled = pd.concat(groups)
skim(base_filled)
```

```python
distritos_viznhos = {
  "Viana do Castelo": ["Braga"],
  "Braga": ["Viana do Castelo", "Vila Real", "Porto"],
  "Vila Real": ["Braga", "Bragança", "Porto", "Viseu"],
  "Bragança": ["Vila Real", "Viseu", "Guarda"],
  "Porto": ["Braga", "Vila Real", "Aveiro", "Viseu"],
  "Aveiro": ["Porto", "Viseu", "Coimbra"],
  "Viseu": ["Vila Real", "Bragança", "Porto", "Aveiro", "Guarda", "Coimbra"],
  "Guarda": ["Bragança", "Viseu", "Coimbra", "Castelo Branco"],
  "Coimbra": ["Aveiro", "Viseu", "Guarda", "Castelo Branco", "Leiria"],
  "Castelo Branco": ["Guarda", "Coimbra", "Leiria", "Santarém", "Portalegre"],
  "Leiria": ["Coimbra", "Castelo Branco", "Santarém", "Lisboa"],
  "Santarém": ["Castelo Branco", "Leiria", "Lisboa", "Setúbal", "Portalegre", "Évora"],
  "Lisboa": ["Setúbal", "Santarém", "Leiria"],
  "Portalegre": ["Castelo Branco", "Santarém", "Évora"],
  "Setúbal": ["Lisboa", "Santarém", "Évora", "Beja"],
  "Évora": ["Portalegre", "Santarém", "Setúbal", "Beja"],
  "Beja": ["Setúbal", "Évora", "Faro"],
  "Faro": ["Beja"]
}
len(distritos_viznhos)
```

```python
# interpolation based on the avg of the neighbors on the same day
from tqdm import tqdm
base_filled.reset_index(inplace=True)
for columns in tqdm(names):
  # get rows with NA
  na_rows = base_filled[base_filled[columns].isna()]
  for index, row in tqdm(na_rows.iterrows()):
    # get the neighbors
    neighbors = distritos_viznhos[row["Estacao"]]
    # get the rows of the neighbors
    neighbors_rows = base_filled[base_filled["Estacao"].isin(neighbors)]
    # get the rows of the neighbors on the same day
    neighbors_rows = neighbors_rows[neighbors_rows["date"] == row["date"]]
    # get the avg of the neighbors
    avg = neighbors_rows[columns].mean()
    base_filled.loc[index, columns] = avg

base_filled[base_filled.isna().any(axis=1)]
```

```python
# transform DD_MED e DD_FFX para sen cos
base_filled["DD_MED_sin"] = np.sin(base_filled["DD_MED"])
base_filled["DD_MED_cos"] = np.cos(base_filled["DD_MED"])
base_filled["DD_FFX_sin"] = np.sin(base_filled["DD_FFX"])
base_filled["DD_FFX_cos"] = np.cos(base_filled["DD_FFX"])
```
