# Zonas

```python
from jsonpath_ng import jsonpath, parse
with codecs.open(here("data/zonasA.json"), "r", "utf-8") as f:
  zonasA = json.load(f)
with codecs.open(here("data/zonasB.json"), "r", "utf-8") as f:
  zonasB = json.load(f)
with codecs.open(here("data/zonasC.json"), "r", "utf-8") as f:
  zonasC = json.load(f)
  
  
# $.results[0].result.data.dsr.DS[0].PH[0].DM0[0].M[0].DM1

path = parse("$.results[0].result.data.dsr.DS[0].PH[0].DM0[0].M[0].DM1[*]")
# agr em for loop
zonas_l = []

for zonas_json in [zonasA, zonasB, zonasC]:
  zonas_json = [match.value["C"] for match in path.find(zonas_json)]
  zonas_json = pd.DataFrame(zonas_json)
  zonas_l.append(zonas_json)
zonas_l[0]["zona"] = "A"
zonas_l[1]["zona"] = "B"
zonas_l[2]["zona"] = "C"
print(zonas_l[0].shape, zonas_l[1].shape, zonas_l[2].shape)
zonas = pd.concat(zonas_l)
zonas = zonas.set_axis(["lat", "long", "drop1", "drop2", "drop3", "zona"], axis=1)
zonas.drop(columns=["drop1", "drop2", "drop3"], inplace=True)
zonas
```

```python
# transform lat e long to string
zonas["lat"] = zonas["lat"].astype(str)
zonas["long"] = zonas["long"].astype(str)
# check where long is not a number and trade lang with long
for i in range(zonas.shape[0]):
  try:
    float(zonas.iloc[i, 1])
  except ValueError:
    zonas.iloc[i, 1], zonas.iloc[i, 0] = zonas.iloc[i, 0], zonas.iloc[i, 1]
    zonas.iloc[i, 0] = zonas.iloc[i-1, 0]

# cast the colums to float
zonas["lat"] = zonas["lat"].astype(float)
zonas["long"] = zonas["long"].astype(float)
zonas.dtypes
```

```python
zonas.value_counts("zona")
```

```python
zonas.to_csv(here("data/zonas.csv"), index=False)
```

```python
import shapely
from shapely.geometry import MultiLineString
import geopandas as gpd
import orjson
import pyogrio
# geopandas.options.io_engine = "pyogrio"
with codecs.open(here("data/network estacoes.json"), "rb", "utf-8") as f:
  network = orjson.loads(f.read())
network = network # sample
pd.DataFrame(network)
```

```python
# build geojson
geojson = {
  "type": "FeatureCollection",
  "features": [
    {
      "type": "Feature",
      "properties": {
        "id": i["geo_digest"]
      },
      "geometry": i["geometry"],
    }
    for i in network
  ] 
}
gdf = gpd.GeoDataFrame.from_features(geojson)
gdf
```

```python
import requests
from time import sleep
from tqdm.notebook import tqdm

gdf["concelho"] = ""
gdf["codigo"] = ""
for i in tqdm(range(promenores.shape[0])):
  i = promenores.iloc[i]
  if i["codigo"] == "":
    continue
  url = "https://e-redes.opendatasoft.com/api/records/1.0/geopreview/?apikey=c0f2ca1314f2e000d94bbcff3726469261d138f10ecc35e6e352d071&clusterprecision=11&dataset=auxiliar2" + "&refine.codigo_se_montante=" + i["codigo"]
  response = requests.get(url)
  if response.status_code == 200:
    response = response.json()
  else:
    raise Exception("Request failed")
  # response: [
  # {
  #   "geo_digest": "1733553459801239980",
  #   "geometry": {
  #     "type": "MultiLineString",
  #     "coordinates": [...
  #     ]
  #   }
  # }, ...
  # ]
  # add to gdf joining by id
  for geo_digest in response:
    gdf.loc[gdf["id"] == geo_digest["geo_digest"], "concelho"] = i["concelho"]
    gdf.loc[gdf["id"] == geo_digest["geo_digest"], "codigo"] = i["codigo"]
```

```python
# dissolve those with the same codigo 
gdf.drop("id", axis="columns", inplace=True)
gdff = gdf.dissolve(by="codigo")
gdff
```

```python
intersections = []
# gdf.iloc[0].geometry.intersects(gdf.iloc[3].geometry)
codigos = list(gdff.index)
for i in tqdm(range(gdff.shape[0])):
  for j in range(i+1, gdff.shape[0]):
    if gdff.iloc[i].geometry.intersects(gdff.iloc[j].geometry):
      intersections.append((codigos[i], codigos[j]))
intersections
```

```python
# intersections is a graph, so calculate the degree of each node
from collections import Counter

nodes = codigos
edges = intersections
degrees = Counter([node for edge in edges for node in edge])
degrees
```

```python
# sum the degrees by concelho
gdff["degree"] = [degrees[i] for i in codigos]
concelho_degress = gdff.groupby("concelho")["degree"].sum()
# drop empty concelho
concelho_degress = concelho_degress[concelho_degress.index != ""]
# sort by degree
concelho_degress.sort_values(ascending=False)
```

```python
# save
gdff.to_file(here("data/network.geojson"), driver="GeoJSON")
concelho_degress.to_csv(here("data/concelho_degress.csv"))
```
