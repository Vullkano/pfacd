```python
# add column with the month and date from id
# codigo_do_relatorio: r"^EREDES|EDPD_(\d{4})_([A-Z]{3})_T|TC_\d+$"
#                                      ano      mes
year_month = duckdb.sql(r"""
  select 
    s.year, 
    s.month 
  from (
    select regexp_extract(codigo_do_relatorio, 'EREDES|EDPD_(\d{4})_([A-Z]{3})_T|TC_\d+$', ['year', 'month']) as s 
    from df
  ) as sub
""").df()
# join them by line number
dff = df.join(year_month)
dff
```

```python
from dateutil.relativedelta import relativedelta as rd
MONTHS = {
  "JAN": 1,
  "FEV": 2,
  "MAR": 3,
  "ABR": 4,
  "MAI": 5,
  "JUN": 6,
  "JUL": 7,
  "AGO": 8,
  "SET": 9,
  "OUT": 10,
  "NOV": 11,
  "DEZ": 12
}

def parse_date(line: pd.Series, formats: list):
  date_string = line["data_do_incidente"]
  for fmt in formats:
    try:
      line["date"] = pd.to_datetime(date_string, format=fmt)
      return line
    except ValueError:
      pass
  
  if type(date_string) is int or date_string.isdigit():
    date = pd.to_datetime(int(date_string), unit="D", origin="1899-12-30")
    # change month to right one
    rightOne = MONTHS[line["month"]]
    date = date + rd(months=rightOne - date.month)
    if int(date.year) != int(line["year"]):
      raise ValueError(f"Year {date.year} is different from {line["year"]}")
    line["date"] = date
    return line
  line["date"] = pd.NaT # Return NaT if none of the formats work
  return line
formats = ["%d/%m/%Y %H:%M:%S", "%d/%m/%Y","%d/%m/%Y %H:%M", "%Y%m%d.0", "%Y-%m-%d", "%d-%m-%Y", "%Y%m%d"]
dfff = dff.apply(axis= "columns", func = lambda x: parse_date(x, formats))
# select where is NaT
dfff[dfff["date"].isna()]
```

```python
concelhos = pd.read_excel(here(r"data/concelhos-metadata.xlsx")) # https://dados.gov.pt/pt/datasets/concelhos-de-portugal/
concelhos
```

```python
import thefuzz.fuzz as fuzz
from thefuzz import process as fuzzy_process
from unidecode import unidecode
concelhos_sem_match = set()
concelhoss = [unidecode(i) for i in concelhos["designacao"].values]

def get_concelho(line: pd.Series):
  concelho = line["concelho_origem_do_incidente"]
  match, score = fuzzy_process.extractOne(unidecode(concelho), concelhoss)
  if score < 95:
    concelhos_sem_match.add((concelho, match, score))
  line["concelho"] = match
  return line

dfffff = dffff.apply(axis="columns", func=get_concelho)
dfffff
```
