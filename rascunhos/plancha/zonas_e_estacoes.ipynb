```python
import json
from pyprojroot.here import here
import codecs

with codecs.open(here("data/instalacoes.json"), "r", "utf-8") as f:
  instalacoes = json.load(f)
instalacoes = [i["instalacao"] for i in instalacoes["results"]]
instalacoes
```

```python
import pandas as pd
with codecs.open(here("data/instalacoes_promenores_1.json"), "r", "utf-8") as f:
  promenores_1 = json.load(f)["results"]
with codecs.open(here("data/instalacoes_promenores_2.json"), "r", "utf-8") as f:
  promenores_2 = json.load(f)["results"]
with codecs.open(here("data/instalacoes_promenores_3.json"), "r", "utf-8") as f:
  promenores_3 = json.load(f)["results"]
with codecs.open(here("data/instalacoes_promenores_4.json"), "r", "utf-8") as f:
  promenores_4 = json.load(f)["results"]
with codecs.open(here("data/instalacoes_promenores_5.json"), "r", "utf-8") as f:
  promenores_5 = json.load(f)["results"]

promenores = promenores_1 + promenores_2 + promenores_3 + promenores_4 + promenores_5
promenores = pd.DataFrame(promenores)
promenores
```

```python
promenores["lon"] = [i["lon"] for i in promenores["coordenadas"]]
promenores["lat"] = [i["lat"] for i in promenores["coordenadas"]]
promenores = promenores.drop(columns=["coordenadas"])
promenores.to_csv(here("data/promenores.csv"), index=False)
```

```python
from skimpy import skim
for col in promenores.select_dtypes(include='object').columns:
  promenores[col] = pd.Categorical(promenores[col])
skim(promenores)
```

```python
promenores = promenores[promenores["tipo_de_instalacao"] != "2025"]
frequencies = promenores.value_counts(["tipo_de_instalacao"])

frequencies
```

```python
from jsonpath_ng import jsonpath, parse
with codecs.open(here("data/zonasA.json"), "r", "utf-8") as f:
  zonasA = json.load(f)
with codecs.open(here("data/zonasB.json"), "r", "utf-8") as f:
  zonasB = json.load(f)
with codecs.open(here("data/zonasC.json"), "r", "utf-8") as f:
  zonasC = json.load(f)
  
  
# $.results[0].result.data.dsr.DS[0].PH[0].DM0[0].M[0].DM1

path = parse("$.results[0].result.data.dsr.DS[0].PH[0].DM0[0].M[0].DM1[*]")
# agr em for loop
zonas_l = []

for zonas_json in [zonasA, zonasB, zonasC]:
  zonas_json = [match.value["C"] for match in path.find(zonas_json)]
  zonas_json = pd.DataFrame(zonas_json)
  zonas_l.append(zonas_json)
zonas_l[0]["zona"] = "A"
zonas_l[1]["zona"] = "B"
zonas_l[2]["zona"] = "C"
print(zonas_l[0].shape, zonas_l[1].shape, zonas_l[2].shape)
zonas = pd.concat(zonas_l)
zonas = zonas.set_axis(["lat", "long", "drop1", "drop2", "drop3", "zona"], axis=1)
zonas.drop(columns=["drop1", "drop2", "drop3"], inplace=True)
zonas
```
